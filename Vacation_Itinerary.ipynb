```python
df = pd.read_csv("weather_data/WeatherPy_vacation.csv")
```

```python
# Filter the DataFrame for each city you want to go to and create separate DataFrames for each city.
yima_df = df.loc[df["City"]=="Yima"]
zhoucheng_df = df.loc[df["City"]=="Zhoucheng"]
dongying_df = df.loc[df["City"]=="Dongying"]
linshui_df = df.loc[df["City"]=="Linshui"]
```

```python
yima = tuple(zip(yima_df["Lat"],yima_df["Lng"]))[0]
zhoucheng = tuple(zip(zhoucheng_df["Lat"],zhoucheng_df["Lng"]))[0]
linshui = tuple(zip(linshui_df["Lat"],linshui_df["Lng"]))[0]
dongying = tuple(zip(dongying_df["Lat"],dongying_df["Lng"]))[0]
```

```python
print(yima,zhoucheng,linshui,dongying)
```

```python
fig = gmaps.figure()
yima2dongying = gmaps.directions_layer(yima, dongying, 
                                       waypoints=[linshui,zhoucheng],
                                       travel_mode='DRIVING')
fig.add_layer(yima2dongying)
fig
```

```python
# Create a new DataFrame that has all the individual city DataFrames created.
hotel_df=yima_df.append(zhoucheng_df).append(linshui_df).append(dongying_df)
```

```python
hotel_df
```

```python
# Create a marker layer map for the four cities

info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description}</dd>
<dt>Max Temp</dt><dd>{Max Temp} Â°F</dd>
</dl>
"""

hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# Add cities to a marker layer map.
locations = hotel_df[["Lat","Lng"]]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

fig.add_layer(marker_layer)
# Call the figure to plot the data.
fig
```
